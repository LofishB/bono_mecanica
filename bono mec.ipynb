```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,100,10000)
w1=1
ri=0.0001

def phi(t):
    return w1*t

phii=phi(t)

def r(phif):
    return (ri/2)*np.exp(phif)+(ri/2)*np.exp(-phif)

x=r(phii)*np.cos(phii)
y=r(phii)*np.sin(phii)
```

```python
##Para omega grande

plt.plot(x,y)
plt.grid()
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.xlim(-1,1)
plt.ylim(-1,1)
```
